### The data inside the data frame is of type (object) 
### And we can't take out the relationship between catjory
### Characteristics must be converted to numbers, and may be done because the values within the columns are an assessment of household conditions

```python
 
final_evaluation = {'not_recom':1,'priority':4,'spec_prior':5,'very_recom':3,'recommend':2}

health = {'not_recom':1,'priority':3,'recommended':2 }

social = {'nonprob':1,'slightly_prob':3,'problematic': 2 }

finance = {'inconv':1,'convenient':2 }

housing = {'convenient':3,'less_conv':2 ,'critical':1}
 
children = {'1':1,'2':2,'3':3,'more':4}

form = {'foster':4,'incomplete':1,'completed':3,'complete':2 }

has_nurs = {'proper':5,'less_proper':4,'improper':3 ,'critical':2,'very_crit':1}

parents = {'usual':2,'pretentious':1,'great_pret':3  }
```

```python
df1 = df.copy()
```

```python
df1['parents'].replace(parents, inplace=True)

df1['has_nurs'].replace(has_nurs, inplace=True)

df1['form'].replace(form, inplace=True)
 
df1['housing'].replace(housing, inplace=True)

df1['finance'].replace(finance, inplace=True)

df1['social'].replace(social, inplace=True)

df1['health'].replace(health, inplace=True)

df1['final evaluation'].replace(final_evaluation, inplace=True)

df1['children'].replace(children, inplace=True)
```

```python
df1
```

```python
df1.corr(min_periods=3)
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df1.corr(method='kendall'), annot=True, fmt='.5f' ,vmin=-1, vmax=1     )
```
